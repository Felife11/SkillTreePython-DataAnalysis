
### **(4) 비교연산**

비교 연산자를 이용하면 boolean 값으로 이루어진 배열을 반환합니다.





#### **1. 값 비교**


|연산자|비교 범용 함수|
|--|--|

|`np.isfinite`|`inf, nan` 값이면 `False`, 아니면 `True`|
|`np.isnan`|`nan` 값이면 `True`, 아니면 `False`|


### **(6) 배열 입출력**



#### **1. `save()`**

넘파이 배열 객체 한개를 파일에 저장합니다.

```
np.save(파일이름, 배열)
```

#### **2. `savez()`**

넘파이 배열 객체 여러개를 파일에 저장합니다.

```
np.savez(파일이름, 배열1, 배열 2, ... 배열n)
```

#### **3. `load()`**

넘파이 배열 저장 파일로부터 객체를 불러옵니다.

```
np.load(파일이름)
```


#### **4. `savetxt()`**

텍스트 파일에 넘파이 배열 객체를 저장합니다. 

```
np.savetxt(파일이름, 배열, delimiter='')
```

#### **5. `loadtxt()`**

텍스트 파일로부터 배열을 불러옵니다.

```
np.loadtxt(파일이름, delimiter=',')
```

####**예제 05**
---

행렬 연산 문제를 넘파이를 이용해서 푸세요

\begin{align*}
 A&= 
\begin{pmatrix}
1 &2  &1 \\ 
2 &1 & 1\\ 
-1 & 0 & 1
\end{pmatrix}
\end{align*}

\begin{align*}
 B&= 
\begin{pmatrix}
1 &0 \\ 
0 &1 \\ 
1 & 0 
\end{pmatrix}
\end{align*}

\begin{align*}
 C&= 
\begin{pmatrix}
1 \\ 
0  \\ 
-1  
\end{pmatrix}
\end{align*}



**05-1**

행렬곱

$A \centerdot B$

**05-2**

$A \centerdot C$

**05-3**

$A * C$

**05-4**

$B \centerdot B^T$

**05-5**

외적

$B \times A$
