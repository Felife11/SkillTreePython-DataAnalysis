### **| 변수 요약**

* 속성을 그리기전 bool 값을 가지는 속성 `adult_mal`과 `alone`을 int 값으로 변환해야 합니다.

```python
df['adult_male'] = df['adult_male'].map({True: 'Yes', False:'No'})
df['alone'] = df['alone'].map({True: 'Yes', False:'No'})
df.head()
```

* 모든 속성의 분포 간단히 그리기

```python
f, axes = plt.subplots(3,5, figsize=(15,10))
axes = axes.flatten()
for col, ax in zip(df.columns, axes):
  sns.histplot(data=df, x=col, ax=ax)
plt.show()
```

* 자료형 정리하기

```python
# 'survived'와 'alive'는 같은 데이터이며 데이터 타입만 다를 뿐.
# `alive`를 예측값으로 사용합니다.
nums = ['age', 'fare']
noms = ['pclass', 'sex', 'sibsp', 'parch', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alone']
y = 'alive'
print('숫자/연속 자료형',nums)
print('범주 자료형', noms)
print('예측할 속성', y)
```

## 데이터 분포
---

### **| 데이터 분포 시각화**

* 연속 자료형 분포 그리기

```python
f, axes = plt.subplots(1, len(nums), figsize=(12,3))
axes = axes.flatten()
for col, ax in zip(nums, axes):
  sns.histplot(data=df, x=col, ax=ax)
plt.show()
```

* 이산 자료형 분포 그리기

```python
f, axes = plt.subplots(2, 6, figsize=(18,8))
axes = axes.flatten()
for col, ax in zip(noms, axes):
  sns.countplot(data=df, x=col, ax=ax)
plt.show()
```

### | **기술 통계량 시각화**

* boxplot 으로 연속 자료형 기술 통계 시각화하기

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 은 시간이 오래 걸리기 때문에 일부 샘플을 추출해 그리는게 좋습니다.
_=sns.pairplot(df.sample(100), hue=y)
```

* `displot` 이용해 연속 자료형 별 예측값의 분포 `kde` 선으로 시각화 하기 

```python
for col in nums:
  sns.displot(data=df, x=col, hue=y, kind='kde')
  plt.show()
```

* 위에서 그린 플롯에 `col` 옵션으로 `sex` 추가하기

```python
for col in nums:
  sns.displot(data=df, x=col, hue=y, col='sex', kind='kde')
  plt.show()
```

* 위에서 그린 플롯에 `hue` 옵션으로 `class` 추가하기

```python
for col in nums:
  sns.displot(data=df, x=col, hue=y, col='class', kind='kde')
  plt.show()
```

* `age`과 `fare`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
_=sns.scatterplot(data=df,x=nums[0], y=nums[1], hue=y)
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `class` 추가하기

```python
_=sns.lmplot(data=df,x=nums[0], y=nums[1], hue=y, col='class')
```

* `histplot`으로 범주형 속성들과 예측값 히트맵 그리기

```python
f, axes = plt.subplots(2,6, figsize=(20,8))
axes = axes.flatten()
for col, ax in zip(noms, axes):
  sns.histplot(data=df, x=col, y=y,ax = ax)
plt.show()
```
