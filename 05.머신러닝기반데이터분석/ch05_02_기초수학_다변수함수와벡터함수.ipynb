### 일변수 스칼라 함수
---

하나의 인풋(=스칼라, 일변수), 하나의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
y = f(x)
\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f(x) :
  return x**2 + 2*x +1
  x = 0
f(0)
````

* 인풋 정의 (-10~10, 100개)

````python
xs = np.linspace(-10, 10, 100)
f(xs)
````

* 2d플롯으로 시각화

````python
#plt.plot(x축, y축)
plt.plot(xs, f(xs))
plt.xlabel('x')
plt.ylabel('f(x)')
````

* 그래프 설명
  * x가 가질수 있는 값 : 무한
  * f(x)가 가질 수 있는 값: 0이상
  * 그래프의 특징 : 최소값이 존재

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03**

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x, y) :
  return x+y
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xs = np.linspace(-15, 20, 18)
ys = np.linspace(-15, 20, 18)
X, Y = np.meshgrid(xs, ys)
print(X, Y)
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize =(9,7))#그림판
ax = plt.axes(projection =  '3d')#축

ax.scatter3D(X, Y, f3(X, Y), marker = '.', c='cyan') #3d 산점도 (x, y, z)
surf = ax.plot_surface(X, Y, f3(X, Y), alpha = 0.6, cmap = 'rainbow' )# 면 surface
fig.colorbar(surf)
#기타 제목/ 축제목 등등..
ax.set_title('f(x,y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 
  * f(x)가 가질 수 있는 값 (아웃풋) : 
  * 그래프의 특징 :

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x):
  h = x**2
  g = 5*x
  return np.array([h, g])

h, g = f6(1)
print(h, g)
````

* 인풋 정의 (-10~10 20개)

````python
xs =np.linspace(-10, 10, 21)
hs, gs = f6(xs)
print(xs)
print(hs)
print(gs)
````

* 2d플롯으로 시각화

````python
# 인풋은 색깔로 표현
# 아웃풋 2개를 2차원에 그리기
plt.plot(gs, hs) # x축값, y축값 선그리기
scat = plt.scatter(gs, hs, c =xs, cmap = 'rainbow') # 점그리기
plt.colorbar(scat)
plt.xlabel('g(x)')
plt.ylabel('h(x)')
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 
  * h(x)가 가질수 있는 값 (두번째 아웃풋): 

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f7(t):
  x = np.sin(6*t)
  y = t/4
  z = (t**2)/2
  return np.array([x,y,z])

f7(1)
````

* 인풋 정의 (-1~3 200개)

````python
ts = np.linspace(-1, 3, 200)
xs,ys,zs = f7(ts)

print(ts. shape, xs.shape, ys.shape, zs.shape)
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize =(9,7))#그림판
ax = plt.axes(projection =  '3d')#축

scat = ax.scatter3D(xs, ys, zs, marker = '.', c=ts) #3d 산점도 (x, y, z)
#surf = ax.plot_surface(xs, ys, zs, alpha = 0.6, cmap = 'rainbow' )# 면 surface
fig.colorbar(scat, shrink =0.6)

#기타 제목/ 축제목 등등..
ax.set_title('f(t) output')
ax.set_xlabel('$sin(6t)$')
ax.set_ylabel('$t/4$')
ax.set_zlabel('$t^2/2$')
plt.show()
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) : 
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) : 

#### **예제 08**

\begin{align}
\tau(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
cos(t)\\
sin(t)\\
t\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def r(t) :
  x = np.cos(t)
  y = np.sin(t)
  z = t
  return np.array([x,y,z])

````

* 인풋 정의(-10 ~ 10 사이 200개)

````python
ts = np.linspace(-10, 10, 200)
xs,ys,zs = r(ts)

print(ts. shape, xs.shape, ys.shape, zs.shape)
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize =(9,7))#그림판
ax = plt.axes(projection =  '3d')#축

scat = ax.scatter3D(xs, ys, zs, marker = '.', c=ts) #3d 산점도 (x, y, z)
#surf = ax.plot_surface(xs, ys, zs, alpha = 0.6, cmap = 'rainbow' )# 면 surface
fig.colorbar(scat, shrink =0.6)

#기타 제목/ 축제목 등등..
ax.set_title('r(t) output')
ax.set_xlabel('$cos(t)$')
ax.set_ylabel('$sin(t)$')
ax.set_zlabel('$t$')
plt.show()
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) :

## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f9(x,y):
  return np.array([x**2, y**2, x*y])

f9(1,2)
````

* 인풋 정의 (x,y 모두 -0.5 ~ 1 사이 50개)

````python
xs = np.linspace(-0.5, 1, 50)
ys = np.linspace(-0.5, 1, 50)
X, Y = np.meshgrid(xs, ys)
X.shape, Y.shape
````

````python
F = f9(X, Y)
F.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize =(9,7))#그림판
ax = plt.axes(projection =  '3d')#축

ax.scatter3D(F[0],F[1],F[2], marker = '.', c='cyan') #3d 산점도 (x, y, z)
surf = ax.plot_surface(F[0], F[1], F[2], alpha = 0.6, cmap = 'rainbow' )# 면 surface
fig.colorbar(surf)
#기타 제목/ 축제목 등등..
ax.set_title('f(x,y) outputs')
ax.set_xlabel('$x^2$')
ax.set_ylabel('$y^2$')
ax.set_zlabel('$xy$')
plt.show()
````

#### **예제 10**


\begin{align}
f(u,v) = \begin{pmatrix}
u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* 함수 생성하기

````python
def f10(u,v):
  return np.array([u,v,1+u**2+v/(1+v**2)])

f10(1,0)
````

* 인풋 정의 (u, v 모두 0 ~ 1 사이 50개)

````python
xs = np.linspace(0, 1, 50)
ys = np.linspace(0, 1, 50)
X, Y = np.meshgrid(xs, ys)
X.shape, Y.shape
````

````python
UV = f10(X, Y)
UV.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize =(9,7))#그림판
ax = plt.axes(projection =  '3d')#축

ax.scatter3D(UV[0],UV[1],UV[2], marker = '.', c='cyan') #3d 산점도 (x, y, z)
surf = ax.plot_surface(UV[0], UV[1], UV[2], alpha = 0.6, cmap = 'rainbow' )# 면 surface
fig.colorbar(surf)
#기타 제목/ 축제목 등등..
ax.set_title('f10(u,v) outputs')
ax.set_xlabel('$u$')
ax.set_ylabel('$v$')
ax.set_zlabel('$1+u^2+v/(1+v^2)$')
plt.show()
````
